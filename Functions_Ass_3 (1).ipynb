```python


"""
 Assignment 3 - Functons

Q 1-  What is the difference between a function and a method in Python?

Ans-  Function:

A function is a block of code that performs a specific task.

It can be called anywhere in your program.

Defined using the def keyword.

Functions can exist independently and are not tied to any specific class or object.

Example:
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))


Method:

A method is a function that is associated with an object.

It is defined inside a class and operates on the data contained within that class.

Methods can only be called on objects or instances of the class in which they are defined.

Like functions, methods are defined using the def keyword.

Example:
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

alice = Person("Alice")
print(alice.greet())



Q-  2. Explain the concept of function arguments and parameters in Python.

Ans- Parameters-
Parameters are the variables listed inside the parentheses in the function definition.

They act as placeholders for the values that will be passed to the function when it is called.

Example:
def add(a, b):
    return a + b

In this example, a and b are parameters.

Arguments-
Arguments are the actual values or data you pass into the function when you call it.

They are assigned to the corresponding parameters defined in the function.

result = add(5, 3)

In this example, 5 and 3 are arguments.


Q -What are the different ways to define and call a function in Python?

Ans-
Different Types are as follows-

 1. Standard Function Definition

Define:
def greet(name):
    return f"Hello, {name}!"

Call:
print(greet("Alice"))

2. Lambda Functions
Lambda functions are small anonymous functions defined with the lambda keyword. They can have any number of arguments but only one expression.
 Define:
add = lambda a, b: a + b

Call:

print(add(5, 3))

3. Recursive Functions
A recursive function is a function that calls itself in order to solve a problem.
 Define:
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)

   Call:
print(factorial(5))  # Output: 120

4. Functions with Default Arguments
Define:
def greet(name, age=25):
    return f"Hello, {name}! You are {age} years old."


5. Functions with Variable-Length Arguments
Define:
def sum_all(*args):
    return sum(args)
Call:
print(sum_all(1, 2, 3, 4))  # Output: 10
Define:
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
Call:
display_info(name="Alice", age=30)

6. Nested Functions
A nested function is a function defined inside another function.
Define:

def outer_function(msg):
    def inner_function():
        print(msg)
    inner_function()
Call:

python
outer_function("Hello from the outer function!")



Q-4 What is the purpose of the `return` statement in a Python function?
Ans-
The return statement in a Python function has a vital role. Its main purposes include:

1- Returning a Value: It allows the function to send a value back to the caller. This value can be of any data type, including numbers, strings, lists, dictionaries,
 or even another function.

def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8

2- Exiting the Function: When a return statement is encountered, the function terminates and exits immediately. Any code after the return statement will not be executed.

def check_even(number)    if number % 2 == 0:
        return True
    return False

print(check_even(4))  # Output: True
print(check_even(5))  # Output: False

3- Returning Multiple Values: Using tuples, you can return multiple values from a function.

def get_name_and_age():
    name = "Alice"
    age = 30
    return name, age

name, age = get_name_and_age()
print(f"Name: {name}, Age: {age}")
In summary, the return statement is crucial for making functions reusable, flexible, and meaningful in your code.



Q 5. What are iterators in Python and how do they differ from iterables?
Ans- Iterables:-
Iterables are objects that can be looped over (iterated over). They include sequences like lists, tuples, strings, and other objects like dictionaries and sets.

An iterable implements the __iter__() method, which returns an iterator.

Examples of Iterables:

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_string = "hello"
my_dict = {'a': 1, 'b': 2}
my_set = {1, 2, 3}

for item in my_list:
    print(item)  # 1 2 3

Iterators:-
Iterators are objects that represent a stream of data. They produce data one element at a time.

An iterator implements two methods: __iter__() (returns itself) and __next__() (returns the next element).

When an iterator is exhausted, it raises a StopIteration exception.

Examples of Iterators:

# Using iter() to create an iterator from an iterable
my_list = [1, 2, 3]
my_iterator = iter(my_list)

print(next(my_iterator))  # 1
print(next(my_iterator))  # 2
print(next(my_iterator))  # 3

Differences-
1) An iterable can be looped over but does not keep track of its current state.

2) An iterator provides a way to access elements sequentially and keeps track of its state (the position of the next element).

Q 6. Explain the concept of generators in Python and how they are defined.

Ans- Generators in Python are a special type of iterator that allow you to generate values on the fly without storing them in memory all at once.
 They are particularly useful for working with large data sets or streams of data where you only need to process one item at a time.

Key Features of Generators:
Lazy Evaluation: Generators produce items only when requested, which makes them memory-efficient.

State Persistence: They maintain state between iterations, allowing them to resume where they left off.

Defining Generators:
Generators can be defined in two ways: using generator functions or generator expressions.

1. Generator Functions:
A generator function is defined using the def keyword and the yield statement. Each call to yield produces a value and pauses the function, maintaining its state for subsequent calls.

Example:
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

# Using the generator function
counter = count_up_to(5)
for number in counter:
    print(number)
Output:

1
2
3
4
5
2. Generator Expressions:
Generator expressions are similar to list comprehensions but use parentheses instead of square brackets. They provide a concise way to create generators.

Example:
squares = (x * x for x in range(1, 6))

# Using the generator expression
for square in squares:
    print(square)
Output:

1
4
9
16
25
How Generators Work:
When a generator function is called, it returns a generator object without executing the function. When next() is called on the generator object,
the function executes until it hits a yield statement. The yield statement returns a value, and the function pauses, saving its state. When next() is called again,
the function resumes where it left off.

Benefits of Generators:
Memory Efficiency: They produce values one at a time, which is more memory-efficient than storing a large list of values.

Improved Performance: They can provide performance improvements when working with large data sets.


Q  7. What are the advantages of using generators over regular functions?

Ans- Using generators in Python offers significant advantages over regular functions, particularly when handling large datasets or streams of data. One key benefit is memory
 efficiency. Generators generate values on demand without storing the entire dataset in memory, making them ideal for large or infinite sequences. This is achieved through
  lazy evaluation, where values are produced only when needed, reducing memory usage and improving performance.

Generators provide improved performance as they yield values one at a time and maintain state between iterations, unlike functions that return entire lists or data structures.
 Additionally, generators simplify code by replacing complex iteration protocols with cleaner, more readable generator functions, making the code easier to understand and maintain.

Another advantage is the ability to create infinite sequences, which is not feasible with regular functions that return finite collections. Generators can also build data
processing pipelines, allowing for modular and efficient data processing. By chaining multiple generators, you can create streamlined workflows that process data in stages,
 enhancing code modularity and reusability.

Example:
python
def infinite_numbers(start=0):
    while True:
        yield start
        start += 1

gen = infinite_numbers()
for _ in range(5):
    print(next(gen))  # Output: 0, 1, 2, 3, 4

In summary, generators offer memory efficiency, lazy evaluation, improved performance, code simplicity, and support for infinite sequences and data pipelines.


Q 8. What is a lambda function in Python and when is it typically used?

Ans- 8. What is a lambda function in Python and when is it typically used? in 150 words with example
A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can have any number of arguments but only one expression.
 Lambda functions are often used for short, simple tasks where defining a full function would be overkill.

When to Use Lambda Functions:
Short and Simple Functions: For small, throwaway functions used within a single line of code.

Sorting and Filtering: Useful with functions like sorted(), filter(), and map() to define short operations.

Inline Function Definitions: When you need a brief function without the overhead of a full definition.

Example:
# Using lambda with sorted()
my_list = [(1, 'one'), (3, 'three'), (2, 'two')]
sorted_list = sorted(my_list, key=lambda x: x[1])
print(sorted_list)  # Output: [(1, 'one'), (2, 'two'), (3, 'three')]

# Using lambda with map()
squared = list(map(lambda x: x * x, [1, 2, 3, 4]))
print(squared)  # Output: [1, 4, 9, 16]


Q 9. Explain the purpose and usage of the `map()` function in Python.

Ans- The map() function in Python is used to apply a specified function to each item of an iterable (e.g., list, tuple) and return a map object (an iterator) containing the results. It's particularly useful for transforming data elements in a clean and efficient way.

Purpose:
The main purpose of map() is to streamline operations that involve applying a function to each item in a sequence without needing to write explicit loops.

Usage:
Applying a Function to a List:
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]

Using Lambda Functions with map():

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x * x, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]

Multiple Iterables: You can also pass multiple iterables to map(), and the function will apply element-wise.

def add(x, y):
    return x + y

numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
result = map(add, numbers1, numbers2)
print(list(result))  # Output: [5, 7, 9]


Q  10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans
 map()
Purpose: Applies a given function to each item in an iterable (e.g., list) and returns a map object (an iterator) of the results.

Usage: Useful for transforming data elements.

Example:

numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]

filter()
Purpose: Applies a given function (which returns a boolean) to each item in an iterable and returns a filter object (an iterator) containing only the items
 for which the function returned True.

Usage: Useful for filtering out elements from a collection based on a condition.

Example:
numbers = [1, 2, 3, 4]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))  # Output: [2, 4]

reduce()
Purpose: Applies a given function to the items of an iterable cumulatively, reducing the iterable to a single value.

Usage: Useful for cumulative operations such as summing or multiplying all elements.

Example:

from functools import reduce

numbers = [1, 2, 3, 4]
total = reduce(lambda x, y: x + y, numbers)
print(total)  # Output: 10

Summary of Differences:
map(): Transforms each element in an iterable.

filter(): Filters elements based on a condition.

reduce(): Aggregates elements to produce a single result.


"""

```

```python
# Q 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    # Initialize an empty list to store the squares
    squared_numbers = []

    # Iterate through the list of numbers
    for num in numbers:
        # Append the square of each number to the new list
        squared_numbers.append(num * num)

    # Return the list of squared numbers
    return squared_numbers

# Example usage
numbers = [1, 2, 3, 4, 5]
print(square_numbers(numbers))  # Output: [1, 4, 9, 16, 25]
```

```python

# Q 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(number):
    # Check if the number is less than 2 (not prime)
    if number < 2:
        return False
    # Check divisibility from 2 to the square root of the number
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False
    return True

# Testing the function on numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)

```

```python
# Q  5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.index = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.num_terms:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return self.a
        elif self.index == 1:
            self.index += 1
            return self.b
        else:
            self.index += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

# Example usage:
num_terms = 10
fibonacci_sequence = Fibonacci(num_terms)
for number in fibonacci_sequence:
    print(number)

```

```python

# Q  6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(max_exponent):
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

# Example usage
max_exponent = 5
for power in powers_of_2(max_exponent):
    print(power)
```

```python

# Q  7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # strip() to remove any leading/trailing whitespace

# Example usage
file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
    print(line)


```

```python

# Q  9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Using map() to apply the conversion function to each temperature in the list
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Printing the converted temperatures
print(fahrenheit_temps)
```

```python

# Q 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    # Use filter() to exclude vowels from the input string
    filtered_string = filter(lambda char: char not in vowels, input_string)
    # Convert the filter object to a string
    result = ''.join(filtered_string)
    return result

# Example usage
input_string = "Hello, world!"
result = remove_vowels(input_string)
print(result)  # Output: "Hll, wrld!"
```
